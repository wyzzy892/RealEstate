```python
import re
import numpy as np
import datetime as dt
```

```python
def extract_year(year, yearbuilding):
    if not np.isnan(year):
        return int(year)
    if type(yearbuilding)==str and yearbuilding.strip()=='сдан':
        return dt.datetime.now().year
    if type(yearbuilding)==str:
        return int(next(iter(re.findall(r"[0-9]{4,4}", yearbuilding)), -1))
    return -1
```

```python
df['год постройки']=df.apply(lambda row: extract_year(row['год постройки'], row['срок сдачи']), axis=1)
```

```python
def extract_height(height):
    if type(height)==str:
        h = float(height.split()[0])
        if h<10:
            return h
        if h>10 and h<100:
            return h/10
        if h>100 and h<1000:
            return h/100
        if h>1000:
            return h/1000
        return h
    return height
```

```python
from functools import reduce
```

```python
categorial_columns_ = list(set(df.columns) - set(df.describe().columns))
```

```python
def contain(row, value):
    if value in row:
        return 1
    else:
        return 0
```

```python
for column_name in categorial_columns_:
    unique_value = reduce(np.union1d, df[column_name].values)
    unique_value = np.delete(unique_value, np.argwhere(unique_value == ''))
    for item in unique_value:
        df[column_name+' '+item] = df[column_name].apply(lambda x: contain(x, item))
    df.drop([column_name], axis=1, inplace=True)
```
