```python
def extract_year(year, yearbuilding):
    if not np.isnan(year):
        year = int(year)
        if year>100:
            return year
        else:
            return (1900+year)
    if type(yearbuilding)==str and yearbuilding.strip()=='сдан':
        return dt.datetime.now().year
    if type(yearbuilding)==str:
        return int(next(iter(re.findall(r"[0-9]{4,4}", yearbuilding)), "19"+next(iter(re.findall(r"[0-9]{2,2}", yearbuilding)), "00")))
    return dt.datetime.now().year
```

```python
# этаж
ldict = dict(df['этаж'].value_counts())
lkeys = ldict.keys()
n = 1
for item in lkeys:
    level = ldict[item]
    if level>400:
        continue
    if 300<=level and level<400:
        n = 2
    elif 200<=level and level<300:
        n = 3
    elif 100<=level and level<200:
        n = 5
    elif 90<=level and level<100:
        n = 6
    elif 80<=level and level<90:
        n = 8
    elif 70<=level and level<80:
        n = 9
    elif 60<=level and level<70:
        n = 10
    elif 50<=level and level<60:
        n = 11
    elif 40<=level and level<50:
        n = 15
    elif 30<=level and level<40:
        n = 20
    elif 20<=level and level<30:
        n = 30
    elif 10<=level and level<20:
        n = 60
    elif 7<=level and level<10:
        n = 100
    else:
        n = 130
    tmp = pd.concat([df[df['этаж']==item]]*n)
    df = pd.concat([df,tmp])
```
